```python
'''
#(Q) Draw a 2-D Scatter plot with color-coding for each flower type/class.
# Here 'sns' corresponds to seaborn.
'''
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Load Iris.csv into a pandas dataFrame.
iris = pd.read_csv("iris.csv")
plt.scatter(iris.sepal_width, iris.sepal_length, edgecolors='r')
plt.xlabel('sepal_width')
plt.ylabel('sepal_length')
plt.title('2-D scatter plot of dataset')
plt.show()
sns.scatterplot(x="sepal_width", y="sepal_length",
                    hue="species", style="species", data =iris)
```
