```python
'''
#(Q) Generate a 2D Densse plot for Iris species with x-axis as petal_length and y-axis as petal_width for iris_setosa
'''
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Load Iris.csv into a pandas dataFrame.
iris = pd.read_csv("iris.csv")
#sns.pairplot(iris)
#g = sns.FacetGrid(iris, col="species")
#g.map(plt.hist, "petal_width", color="m")
#sns.boxplot(x="species", y="petal_length",
#             data=iris, linewidth=2.5)
#sns.violinplot(x="species", y="petal_length", data=iris)
setosa = iris.loc[iris.species == "setosa"]
sns.kdeplot(setosa.petal_length, setosa.petal_width,
              cmap="Reds", shade=True, shade_lowest=False)
```
